```python
import glob 
import yaml
import sys
sys.path.append('../../src')

from Environments import *

OBS_KEY = 'obs'
STARTS_KEY = 'starts'
GOALS_KEY = 'goals'
```

```python
import os 

path = '.' # dir with yaml files
names = "*.yaml"
files = glob.glob(os.path.join(path, names))

envs = {}
env_counter = 0
for filename in files:
    print(filename)
    with open(filename, 'r') as file:
        content = file.read()
        filedata = yaml.safe_load(content)
        # print(filedata)

        this_env = {}
        obs = []
        for given_obs in filedata['environment']['obstacles']:
            obs.append((given_obs['center'][0], given_obs['center'][1], given_obs['size'][0],  given_obs['size'][1]))
        this_env[OBS_KEY] = obs

        starts = []
        goals = []
        for given_rob in filedata['robots']:
            given_start = given_rob['start']
            starts.append((given_start[0], given_start[1], given_start[2]))
            given_goal = given_rob['goal']
            goals.append((given_goal[0], given_goal[1]))

        this_env[STARTS_KEY] = starts
        this_env[GOALS_KEY] = goals 

        envs[filename.split('.')[1][1:]] = (this_env)  
                
```

```python
with open("dcbs_envs.py", 'w') as f:
    f.write("from Environments import *\n")
    f.write("OBS_KEY = 'obs'\n")
    f.write("STARTS_KEY = 'starts'\n")
    f.write("GOALS_KEY = 'goals'\n\n")

    f.write("envs = {}\n\n")

    envcounter = 0
    for filename, env in envs.items():
        f.write("# " + filename + "\n")

        f.write("env" + str(env_counter) + " = {}\n")

        f.write("starts" + str(env_counter) + " = [\n")
        for start in env[STARTS_KEY]:
            f.write("\t(" + str(start[0]) + ", "  + str(start[1]) + ", " + str(start[2]) + "),\n")
        f.write("]\n\n")

        f.write("goals" + str(env_counter) + " = [\n")
        for goal in env[GOALS_KEY]:
            f.write("\t(" + str(goal[0]) + ", "  + str(goal[1]) + "),\n")
        f.write("]\n\n")

        f.write("obstacles" + str(env_counter) + " = [\n")
        for obs in env[OBS_KEY]:
            f.write("\tRectangleObstacle2D(" + str(obs[0]) + ", "  + str(obs[1]) + ", " + str(obs[2]) + ", " + str(obs[3]) + "),\n")
        f.write("]\n\n")

        f.write("env" + str(env_counter) + "[STARTS_KEY] = starts" + str(env_counter) + "\n")
        f.write("env" + str(env_counter) + "[GOALS_KEY] = goals" + str(env_counter) + "\n")
        f.write("env" + str(env_counter) + "[OBS_KEY] = obstacles" + str(env_counter) + "\n")
        f.write("envs[\"" + filename + "\"] = env" + str(env_counter) + "\n\n\n")
        env_counter+=1
        
        


```
